```python
from pathlib import Path
import json
import yaml
import torch
import matplotlib.pyplot as plt

from project_main.tokens import build_vocab
from project_main.data import make_batch, make_eval_batch
```

```python
CONFIG_PATH = Path("../configs/base.yaml")
with open(CONFIG_PATH, "r") as f:
    cfg = yaml.safe_load(f)
```

```python
vocab = build_vocab(cfg['task'])
print(vocab.size)
print(vocab.newline_id)
print(vocab.decode(list(range(vocab.size))))
```

```python
batch = make_batch(task_cfg=cfg['task'], vocab=vocab, batch_size=1,device="cpu")
for i in range(1):
    print(f"Example {i}:")
    print("Input: ", vocab.decode(batch.tokens[i].tolist()))
    print("Target: ", vocab.decode(batch.targets[i].tolist()))
    print()
```
