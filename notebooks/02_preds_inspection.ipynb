```python
import importlib
from pathlib import Path
import json
import yaml
import torch
import matplotlib.pyplot as plt

import project_main.tokens as tokens
from project_main.data import make_batch, make_eval_batch
from project_main.model import build_model
from project_main.checkpoints import load_checkpoint

```

```python
RUN_DIR = Path("../results/line_breaks_results")
CHECKPOINT_PATH = RUN_DIR / "checkpoints" / "final.pt"
with open(RUN_DIR / "config.json", "r") as f:
    cfg = json.load(f)

device = "cpu"
vocab = tokens.build_vocab(cfg['task'])
model = build_model(cfg=cfg, device=device)

checkpoint = load_checkpoint(path=CHECKPOINT_PATH, model=model,map_location=device)
model.eval()

checkpoint['step']
```

```python
for i in range(10):
    batch = make_batch(task_cfg=cfg['task'], vocab=vocab, batch_size=1, device=device)
    with torch.no_grad():
        logits = model(batch.tokens)
        predictions = torch.argmax(logits, dim=-1)
print(f"Example: {i}")
# print("Input: ", vocab.decode(batch.tokens[0].tolist()))
print("Targs: ", vocab.decode(batch.targets[0].tolist()))
print("Preds: ", vocab.decode(predictions[0].tolist()))
```
